## Credit Risk

Credit risk is one of the most important variables in the financial system of today. Before we start let's discuss about what is credit risk and get familar with speacial vacobulary which is used in this field.

Credit risk is the likelihood that a debator does not repay their debt.

Measuring credit risk well is the most effective way of ensuring that credit losses would be minimized. Building and implementing the best possible credit risk models would result in the most prediction about expected loss. The lender(Bank)
can also require collaterals. Thous, if borrower does not repay, the assest that serves as a collateral can be acquired by the lender and sold to a third party in order to cover some of the losses in case the borrower defaults. Third, the lender can make sure that the greater loss they expect, the higher the price of lending that they charge. 

There is a certain amount of credit risk associated with every borrower.
Types of factors for Expected loss
    1. Borrower-specific factors
    2. The economic environment 
    3. The combibation of the 1 and 2
    
Second type of credit risk is Unexpected losses it can be result of adverse economic circumstances.

Last type of losses Exceptional it is highly unlikely but it may occure and it may be cause of severe economic downturn.

We will focus way to estimate Expected Loss 

Expected Loss difines product of three frameworks PD(Probability of Default) LGD(Loss given default) EAD(Exposure at default)

$EL = PD*LGD*EAD$ 

Probability of default(PD) - The borrowers inability to repay their debt in full or on time

Loss given default(LGD) - The proportion of the total exposure that cannot be recovered by the lender once a default has occurred

Exposure at default(EAD) - The total value that a lender is exposed to when a borrower defaults

# Capital Adequacy, Regulations and Basel 2 Accord

Order to prevent great consecuencias goverment regulator impose requirements on bank , banks must conduct their business without risking the stability of the economic system

Regulators rules:

    1. Regulate bank operaions and hence reduce risky behavior
    2. Guarantee to the public that the banking system is in good health
    
Banks required to hold sufficient capital that would allow them to absorb the loss from this and other defaults this obligation
This called capital requirment it is also known as capital adequacy or regulatory capital.

Capital requirement(CR)

    Capital/Risk-weighted assets(Loans) >= 8%    <----   Capital adequacy ratio (CAR)

The most important documents that prescribes how much capital bank need to have how capital is defined and how it is compared against risk weighted assests is Basel 2 accords

Basel 2 Accord consists how much capital banks need to have, how capital is defined, how capital is compared against risk-weighted asstes

# Basel 2 Approaches

So far we have looked components of expected loss and we know that it is product PD, LGD, EAD.

Expected Loss 
$$EL = PD*LGD*EAD $$

Basel 2 accord prescribes that regulators should allow banks choose from three different approaches for calculating or modeling credit risk, that is they can choose from one of three different approaches for calculating or modeling each of three different approaches 
    1.Standardized Approach(SA)
    2.Internal Ratings Based(IRB) Approaches
        2.1 Foundation Internal Ratings Based(F-IRB) Approach
        2.2 Advanced Internal Ratings Based(A-IRB) Approach
        
Under the standardized approach(SA), banks are required to use data from external credit agencies to assess credit risk.
Such data provided by agencies FICO credit score for individuals and households, S&P, Moody's and Fitch

### Preprocessing Discrete Variables: Automating Calculaions

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

```python
# WoE function for ordered discrete and continuous variables
def woe_ordered_continuous(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    #df = df.sort_values(['WoE'])
    #df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
